```python
from wordmesh import Wordmesh
```

Open the text file

```python
with open('Jobs-speech.txt') as file:
    text = file.read()
```

Create and plot wordmesh

```python
wm = Wordmesh(text)
wm.plot()
```

Change the clustering criteria to 'cooccurence'. This clusters together words that tend to occur together in the text.

```python
wm = Wordmesh(text, num_keywords=90)
wm.set_clustering_criteria('cooccurence')
wm.set_fontcolor('clustering_criteria')
wm.plot()
```

Change the clustering algorithm.

```python
wm.set_clustering_criteria('cooccurence', clustering_algorithm='TSNE')
wm.plot()
```
